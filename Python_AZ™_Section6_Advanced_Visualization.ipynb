# Coordinates and Diagonals

```python
# Controlling axes and adding diagonals
fg=sns.FacetGrid(movie, row='Genre', col='Year', hue='Genre')
kws= dict(s=50, linewidth=0.5, edgecolor='black')
fg=fg.map( plt.scatter, 'Rotten_Ratings', 'Audience_Ratings', **kws)
fg.set(xlim=(0,100), ylim=(0,100))
for ax in fg.axes.flat:
    ax.plot((0, 100), (0, 100), c='gray', ls='--')
fg.add_legend()
```

## Building Dashboards in Python

```python
sns.set_style("darkgrid")
f, axes= plt.subplots(2,2, figsize=(15,15))

vis5=sns.kdeplot(x=movie.Budget_million, y=movie.Rotten_Ratings, ax=axes[0,0])
#vis4=sns.kdeplot(x=movie.Budget_million, y=movie.Audience_Ratings, ax=axes[0,1])
z=sns.violinplot(data=movie,x='Genre', y='Rotten_Ratings',ax=axes[1,0])
# two plot on 1,1 plase
vis3=sns.kdeplot(x=movie.Rotten_Ratings,y=movie.Audience_Ratings, cmap='Reds', shade=True, ax=axes[1,1])
vis3b=sns.kdeplot(x=movie.Rotten_Ratings,y=movie.Audience_Ratings, cmap='Blues', ax=axes[1,1])
# we use seborn plots, and if we can use another plots, like this:
#plot4= plt.hist(movie.Budget_million)
# we need to do spesial thinks: axes[0,1]
axes[0,1].hist(movie.Budget_million)
```

## Styling Dashboards Tips

```python
sns.set_style("dark", {"axes.facecolor" : "black"})
f, axes= plt.subplots(2,2, figsize=(15,15))

#Plot[0,0]
vis5=sns.kdeplot(x=movie.Budget_million, y=movie.Rotten_Ratings, ax=axes[0,0], cmap='inferno', shade=True, shade_lowest=True)
vis5b=sns.kdeplot(x=movie.Budget_million, y=movie.Rotten_Ratings, cmap='cool', ax=axes[0,0])
#vis4=sns.kdeplot(x=movie.Budget_million, y=movie.Audience_Ratings, ax=axes[0,1])

#Plot[1,0]
z=sns.violinplot(data=movie,x='Genre', y='Rotten_Ratings',ax=axes[1,0], palette='cool')

#Plot[1,1]
# two plot on 1,1 plase
vis3=sns.kdeplot(x=movie.Rotten_Ratings,y=movie.Audience_Ratings, cmap='Blues_r', shade=True, ax=axes[1,1])
vis3b=sns.kdeplot(x=movie.Rotten_Ratings,y=movie.Audience_Ratings, cmap='gist_gray_r', ax=axes[1,1])

#Plot[0,1]
# we use seborn plots, and if we can use another plots, like this:
#plot4= plt.hist(movie.Budget_million)
# we need to do spesial thinks: axes[0,1]
axes[0,1].hist(movie.Budget_million)
```

## Finishing Touches

```python
f, axes= plt.subplots() # like for subplots, but with one plot!
f.set_size_inches(11.7, 8) #Size for A-4 paper
sns.set_style("whitegrid")
list1= list()
for gen in movie.Genre.cat.categories:
    list1.append(movie[movie.Genre == gen].Budget_million)
    mylabels.append(gen)
plt.hist(list1, bins= 30, stacked=True, label=mylabels)
plt.title("Movie Budget", fontsize=35, color="DarkBlue")
plt.xlabel("Budget", fontsize=15, color="Green")
plt.ylabel("Number of Movies", fontsize=15, color="Red")
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.legend(frameon=True,fancybox=True, shadow=True, framealpha=0.5,  prop={'size': 20})
plt.show()
```
